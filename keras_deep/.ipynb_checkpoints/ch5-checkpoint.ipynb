# 머신러닝은 일반화와 최적화 사이의 줄다리기!

그 줄다리기를 잘하는 기준은 과적합/과소적합을 파악하기!

- **과대적합이 일어나는 경우**   
잡음 섞인 데이터  
불확실한 특성  
가짜 상관관계 

```python
from tensorflow.keras.datasets import mnist
import numpy as np

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype("float32") / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype("float32") / 255


train_images_with_noise_channels = np.concatenate(
    [train_images, np.random.random((len(train_images), 784))], axis=1)

train_images_with_zeros_channels = np.concatenate(
    [train_images, np.zeros((len(train_images), 784))], axis=1)


#784 + 784 = 1568
train_images_with_noise_channels.shape
```

```python
#잡음을 섞고 train해보기.

from tensorflow import keras
from tensorflow.keras import layers

def get_model():
    model = keras.Sequential([
        layers.Dense(512, activation="relu"),
        layers.Dense(10, activation="softmax")
    ])
    model.compile(optimizer="rmsprop",
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

model = get_model()
history_noise = model.fit(
    train_images_with_noise_channels, train_labels,
    epochs=10,
    batch_size=128,
    validation_split=0.2)

model = get_model()
history_zeros = model.fit(
    train_images_with_zeros_channels, train_labels,
    epochs=10,
    batch_size=128,
    validation_split=0.2)
```

```python
#에폭에 따른 모델의 검증 정확도 변화 모니터링

import matplotlib.pyplot as plt
val_acc_noise = history_noise.history["val_accuracy"]
val_acc_zeros = history_zeros.history["val_accuracy"]
epochs = range(1, 11)
plt.plot(epochs, val_acc_noise, "b-",
         label="Validation accuracy with noise channels")
plt.plot(epochs, val_acc_zeros, "b--",
         label="Validation accuracy with zeros channels")
plt.title("Effect of noise channels on validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
# 랜덤하게 섞은 레이블로 학습
(train_images, train_labels), _ = mnist.load_data()
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype("float32") / 255

#레이블을 랜덤하게 섞기
random_train_labels = train_labels[:]
np.random.shuffle(random_train_labels)

model = keras.Sequential([
    layers.Dense(512, activation="relu"),
    layers.Dense(10, activation="softmax")
])
model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
history = model.fit(train_images, random_train_labels,
          epochs=30,
          batch_size=128,
          validation_split=0.2)
```

```python
accuracy = history.history["accuracy"]
val_acc = history.history["val_accuracy"]
epochs = range(1, 31)
plt.plot(epochs, accuracy, "b-",
         label="train accuracy ")

plt.plot(epochs, val_acc, "b--",
         label="Validation accuracy")
         
plt.title("acc vs val_acc")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
model.evaluate(test_images, test_labels)
```

레이블을 막 섞어서 훈련해도 acc는 올라간다.  
반면 val, test acc는 현저이 낮다.  
결국 일반화의 본질은 딥러닝 모델 자체와는 관련이 없는 것이다!

### 그럼 일반화의 본질은 무엇일까?

-> 매니폴드 가설
-> 보간

이 두 개념이 존재한다. 데이터 포인트가 존재하는 부모 공간에서 매니폴드라는 보다 저차원 공간으로 입력 공간을 학습하면 된다고 한다는데  
무슨 소린지 모르겠다..  
  
그리고 보간이라는 것은 매니폴드의 보간과 선형 보간이 있는데 선형 보간은 부모 공간에서의 평균을 의미한다.  
반면 매니폴드 보간은 저차원의 매니폴드에서의 중간 포인트를 의미한다.  
내가 이해하기로는 선형 보간은 실제 부모차원에서의 값의 평균이고,  
매니폴드 보간은 우리가 이해한 값의 모양대로 모델이 이해하는 값의 중앙이라고 생각했는데  
말로 표현해도 무슨 말인지 모르겠다 ..  

**결론: 근사적으로 학습된 데이터 매니폴드에서 보간을 통해 딥러닝의 일반화가 달성된다!(전부는 아님)**

### 일반화 성능 항샹시키기
데이터셋을 확인하자: 잡음이 많거나 불연속적인 경우 일반화 성능이 떨어진다.(잠재 매니폴드를 생각해보자)  
    - 데이터가 충분한가 
    - 레이블 할당 에러를 최소화하자  
    - 특성 선택의 중요성  
    - 특성 공학 사용  

```python
# 특성 공학: 모델이 수월하게 작업할 수 있는 어떠한 방식으로 데이터를 표현하자(하드코딩)
# 모델 규제: 일단 과대적합과 과소적합의 사이의 성능을 보이는 즉, 적절하게 학습할 수 있는 모델의 크기를 실험적으로 찾자!(with 검증 세트)

# 모델의 크기로 과대/과소 확인하기
from tensorflow.keras.datasets import imdb
import numpy as np
from tensorflow.keras import Sequential, layers

(train_data, train_labels),_ = imdb.load_data(num_words = 10000)

def vectorize_seq(seq, dim = 10000):
    results = np.zeros((len(seq), dim))
    for i, seq in enumerate(seq):
        results[i,seq] = 1.
    return results

train_data = vectorize_seq(train_data)


model = Sequential([
    layers.Dense(16, activation="relu"),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
history_original = model.fit(train_data, train_labels,
                             epochs=20, batch_size=512, validation_split=0.4)


model = Sequential([layers.Dense(4, activation="relu"),
                   layers.Dense(4, activation = 'relu'),
                   layers.Dense(4, activation = 'sigmoid')])

model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history_smaller_model = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)


model = Sequential([layers.Dense(512, activation="relu"),
                   layers.Dense(512, activation = 'relu'),
                   layers.Dense(1, activation = 'sigmoid')])

model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history_larger_model = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt

val_loss1 = history_original.history["val_loss"]
val_loss2 = history_smaller_model.history["val_loss"]

epochs = range(1, 21)
plt.plot(epochs, val_loss1, "b--",
         label="Validation loss of og model")
plt.plot(epochs, val_loss2, "r--",
         label="Validation loss of smaller model")
plt.title("Effect of insufficient model capacity on validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
val_loss1 = history_original.history["val_loss"]
val_loss2 = history_larger_model.history["val_loss"]

epochs = range(1, 21)
plt.plot(epochs, val_loss1, "b--",
         label="Validation loss of og model")
plt.plot(epochs, val_loss2, "r--",
         label="Validation loss of larger model")
plt.title("Effect of insufficient model capacity on validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

### 가중치의 규제

- 드롭아웃 기법

```python
#L2 규제 예시

from tensorflow.keras import regularizers

model = Sequential([
    layers.Dense(16,
                 kernel_regularizer=regularizers.l2(0.002),
                 activation="relu"),
    layers.Dense(16,
                 kernel_regularizer=regularizers.l2(0.002),
                 activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
history_l2_reg = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)
```

```python
#드롭아웃 기법 추가하기: imdb

model = Sequential([
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

history_dropout = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)
```
